```python
import matplotlib.pyplot as plt
import librosa, librosa.display
import numpy as np
import IPython.display as ipd
```

```python
ipd.Audio('piano_middle_C.wav')
```

```python
signal, sampling_rate = librosa.load('piano_middle_C.wav')
```

```python
print(f'no of samples {len(signal)}')
print(f'sampling rate {sampling_rate}')
```

### Short Time Fourier Transform

```python
signal_stft = librosa.stft(signal, n_fft=2048, hop_length= 512)
```

```python
signal_stft.shape
```

### Spectogram

```python
# Square Magnitude of the stft coefficient
signal_spec = np.abs(signal_stft) ** 2
```

```python
signal_spec.shape
```

### Visualize

```python
def plot_spectogram(signal_spec, sampling_rate ,hop_size, y_axis= 'linear'):
    plt.figure(figsize=(25,10))
    librosa.display.specshow(signal_spec, sr= sampling_rate,hop_length=hop_size,
                             x_axis='time', y_axis= y_axis)
    plt.colorbar(format='%+2.f')
```

```python
plot_spectogram(signal_spec, sampling_rate=sampling_rate, hop_size=512)
```

### Log Amplitude Spectogram
* Move the amplitude from the linear to the logarthimic -- that is the way we perceive the audio *

```python
signal_spec_amlog = librosa.power_to_db(signal_spec)
```

```python
signal_spec_amlog.shape
```

```python
plot_spectogram(signal_spec_amlog, sampling_rate= sampling_rate, hop_size= 512)
```

### Log Frequency Spectogram
### Final Result

```python
plot_spectogram(signal_spec_amlog, sampling_rate= sampling_rate, hop_size= 512, y_axis='log')
```
